```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("../journal/journal.csv")
```

```python
df
```

```python
x = np.arange(len(df["Fecha"]))
ancho = 0.35  # Ancho de las barras

plt.figure(figsize=(8, 5))
plt.bar(x - ancho/2, df["Aprobados"], width=ancho, label='Aprobados', color='red')
plt.bar(x + ancho/2, df["Desaprobados"], width=ancho, label='Reprobados', color='green')

# Etiquetas y título
plt.xlabel('Fecha')
plt.ylabel('Cantidad')
plt.title('Comparación de Cantidad por Fecha')
plt.xticks(x, df["Fecha"])
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Guardar imagen
plt.tight_layout()
plt.savefig('comparacion_ventas.png', dpi=300)

# Mostrar
plt.show()
```
